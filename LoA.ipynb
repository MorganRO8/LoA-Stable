# How to Create a Custom Job

## Interactive Mode

Next up, we need to define a job! There is a test example set up, so if you don't feel like coming up with something yourself you can absolutely just skip this section and move on to that!

To make this as easy as possible, there are a few different routes we can take. First, we can start the code in interactive mode and go through the prompts to define the structure we want in our output csv, and the rest of the settings needed to run a job.

````python
!python main.py
````

You'll need to go through the options in that menu to run a job. Option 1 defines the structure of the output csv for the model to match. Option 2 defines the settings for the job, such as what keywords to use when searching for papers, how many to get, etc. Option 3 lets you start a job if you already have these two files.

## Manually Editing Input Files


Let's take a look at the contents of these files we're creating. If we understand the structure, we can write them ourselves!

First, let's look at our example.pkl, which is the csv schema:

````python
!cat dataModels/example.pkl
````

As you can see, this is a simple example. It is meant primarily to show off the structure of the file. In this example, the first column is the name of a chemical, then we have the melting point, boiling point, and notes as the subsequent columns. We define types for each column which helps to filter out incorrect responses. We also define a description for each column, to help the model understand exactly what it is you are looking for. We did our best to make the structure of this file human readable, that way if you want to you could simply make a copy of this file and edit it by hand to define a new schema.

Let's take a look at the job settings for our example:

````python
!cat job_scripts/example.json
````

The job settings file is a bit less readable, admittedly. This helps ensuring that older files work with newer versions, and still the json file can be manageable edited by the user to set up a new job. This example shows all fields which are relevant to include, though not all of them need to be included, and will simply use default values if none are provided. In this example, we are only searching for a few papers, and only from pubmed and arXiv. This is due to colab having issues with the other scraping functions. If you install LoA on your local machine, you can enable those.

Otherwise, the main things to look at when editing this file are the search terms. This is your first line of defense against irrelevant data. Ensuring you select proper terms is imperative. It can be tricky to know which combination will be best, so if time isn't an issue, you can use the 'maybe' search terms as we have here. This will result in a combinatorial creation of multiple searches. Note that 'retmax', which is the max number of papers to download, is enforced per search. So, when using maybe search terms, this limit may be far exceeded overall.


## Lazy Mode

There is one final way to generate job files, and that is using AI assistance. In the next cell, we'll ask you for an idea of what you would like to extract, and the model will do its best to create two new files generated.pkl and generated.json, which you can use for your extraction task. Obvious disclaimer: it may make mistakes, so check through the files it makes if you want to be sure they are correct.

Tip: If you can't think of something, IC50 values of novel cancer treatments is a good one to try!

````python
!nohup ollama serve >/dev/null 2>&1 &
!ollama pull deepseek-r1:8b

import requests
import re
import json

def generate_chemical_files():
    # 1) Ask user what type of chemical data they'd like to extract
    user_input = input("What kind of chemical data do you want to extract? ")

    # 2) Read example files (job script and schema)
    with open('job_scripts/example.json', 'r') as f:
        example_json = f.read()

    # If example.pkl is truly a binary pickle, you'd normally use pickle.load().
    # Here we assume it's text (for demonstration).
    with open('dataModels/example.pkl', 'rb') as f:
        example_pkl_bytes = f.read()
    example_pkl_text = example_pkl_bytes.decode('utf-8', errors='replace')

    # 3) Construct the prompt clearly specifying the desired format
    prompt = f"""
We are building a demonstration for how to generate new job settings (JSON) and data model (text) files for a chemical data extraction pipeline.

The user wants to extract the following type of chemical data: {user_input}

IMPORTANT:
- Provide the job script under a heading exactly labeled: ### job script
- Do NOT include ```json or ```python in your response
- Provide the data model under a heading exactly labeled: ### data model
- Keep "Key Columns: 1" unchanged (this identifies the IUPAC name).
- Do NOT include additional commentary or <think> blocks.
- The files will be saved as generated.json and generated.pkl, respectively, so when linking the .pkl in the .json, use generated.pkl
- For the output csv, pick a relevant file name and put it in the results folder.
- The def search terms should be general/broad.
- When designing search terms, think of keywords that may pull up relevant papers.
- Do NOT use any maybe search terms, instead leave a blank string ('') or 'none'.
- We want to concurrently scrape pubmed and arxiv, but not the others.
- Keep retmax low for now.
- Pick logical data types for each column in the schema.
- We currently support: int, str, boolean, float, range, and complex.
- Replace the existing columns (melting point and boiling point) instead of adding on to them, unless the user requests this.
- As with the above point, utilize the examples given as examples of the necessary structure of the files, not necessarily as to what to include.
- Again, do not include the melting or boiling point columns unless the user requests this as one of the things they'd like to extract.
- Absolutely DO NOT change what models are being used. We must use falcon3:10b for check, and deepseek-r1:8b for extraction in our current environment.

Here is the current job script:
```
{example_json}
```

Here is the current data model:
```
{example_pkl_text}
```

Please update both files accordingly and output exactly as follows:

### job script
<updated JSON here>

### data model
<updated data model text here>
"""

    # 4) Send request to Ollama (streaming)
    url = "http://localhost:11434/api/generate"
    payload = {
        "model": "deepseek-r1:8b",
        "prompt": prompt,
        "temperature": 0.2,
        "stream": True
    }
    headers = {"Content-Type": "application/json"}

    response = requests.post(url, json=payload, headers=headers, stream=True)
    if response.status_code != 200:
        print("Error calling model:", response.text)
        return

    # 5) Stream the response, printing partial output
    print("Model is generating... (streaming partial output below)\n")
    streamed_text = []

    for line in response.iter_lines(decode_unicode=True):
        if not line.strip():
            continue
        try:
            data = json.loads(line)
        except json.JSONDecodeError:
            continue

        chunk = data.get("response", "")
        done = data.get("done", False)

        if chunk:
            print(chunk, end="", flush=True)
            streamed_text.append(chunk)

        if done:
            break

    print("\n\nModel has finished generating.\n")

    # Combine everything into a single text
    model_output = "".join(streamed_text)

    # 6) Extract updated file contents using clearly defined headers
    job_script_pattern = r"### job script\s*(.*?)\s*(?=### data model)"
    data_model_pattern = r"### data model\s*(.*)"

    job_match = re.search(job_script_pattern, model_output, flags=re.DOTALL)
    data_match = re.search(data_model_pattern, model_output, flags=re.DOTALL)

    if not job_match or not data_match:
        print("Unable to parse the returned text. Check the model output or your regex patterns.")
        return

    updated_job_script = job_match.group(1).strip()
    updated_data_model = data_match.group(1).strip()

    updated_job_script = updated_job_script.replace("```json", "").replace("```", "").replace("example.pkl","generated.pkl")

    # 7) Write updated content to new files
    with open('job_scripts/generated.json', 'w') as f:
        f.write(updated_job_script)

    with open('dataModels/generated.pkl', 'w') as f:
        f.write(updated_data_model)

    print("Generated files saved as:")
    print("  • job_scripts/generated.json")
    print("  • dataModels/generated.pkl")

# Run the function
generate_chemical_files()

# Make sure ollama is running
!nohup ollama serve >/dev/null 2>&1 &

# Go!
!python main.py -auto /content/LoA/job_scripts/generated.json
````
